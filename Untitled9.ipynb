```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('bank-full.csv')
```

```python
#spliting
def spll(y,i):
  return y.split(';')[i]
#remove ""
def rem(x):
  s=pd.Series(x)
  s=s.str.replace('"', '')
  return s
```

```python
a=df.columns[0].split(';')
a=rem(a)
```

```python
d={}
for i in range(len(a)):
  d[a[i]]=df.iloc[:,0].apply(lambda x:spll(str(x),i))
```

```python
df1=pd.DataFrame(d)
df1.head()
```

```python
for n in range(len(a)):
  df1.iloc[:,n]=df1.iloc[:,n].apply(lambda x: rem(x))

```
